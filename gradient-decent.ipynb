```python
import numpy as np
import matplotlib.pyplot as plt

def f_cubic(x):
    return (x ** 3) - (7 * x) + 6

def df_cubic(x):
    return 3*(x**2) - 7

def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    trajectory = [x]

    for _ in range(num_iterations):
        grad = df_cubic(x)
        x -= learning_rate * grad
        trajectory.append(x)

    return x, trajectory

starting_point = 16
learning_rate = 0.01
num_iterations = 50

x = np.arange(-3, 4, 0.1)
y = f_cubic(x)

final_x, trajectory_x = gradient_descent(starting_point, learning_rate, num_iterations)
trajectory_y = f_cubic(np.array(trajectory_x))

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^3 - 7x + 6', color='blue')
plt.axhline(0, color='black',linewidth=0.7, ls='-')
plt.axvline(0, color='black',linewidth=0.7, ls='-')

plt.plot(trajectory_x, trajectory_y, marker='o', color='red', label='Gradient Descent Path')

plt.title('Plot of the Cubic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.ylim(-10, 20)
plt.xlim(-3, 4)
plt.show()

print(f'Starting x after gradient descent: {trajectory_x[0]}')
print(f'Final x after gradient descent: {final_x:.4f}')

print(f'Starting y after gradient descent: {trajectory_y[0]}')
print(f'Final y after gradient descent: {trajectory_y[-1]}')
```
