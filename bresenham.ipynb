```python
import numpy as np
import matplotlib.pyplot as plt

def bresenham(x1, y1, x2, y2):
    # Initialize the list of points
    points = []
    
    # Calculate the differences between the two points
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    # Determine the direction of the increments (either +1 or -1)
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1
    
    # Initialize the error term
    err = dx - dy
    
    # Loop until we reach the end point
    while True:
        # Append the current point
        points.append((x1, y1))
        
        # Check if we have reached the end point
        if x1 == x2 and y1 == y2:
            break
        
        # Calculate the error term for the next step
        e2 = 2 * err
        
        # Adjust the error term and the x or y coordinates
        if e2 > -dy:
            err -= dy
            x1 += sx
        if e2 < dx:
            err += dx
            y1 += sy
    
    return np.array(points)

# Example of using the algorithm
x1, y1 = 2, 3
x2, y2 = 10, 7
points = bresenham(x1, y1, x2, y2)

# Plot the result using Matplotlib
grid = np.zeros((15, 15))  # Create a grid for visualization

# for point in points:
#     grid[point[1], point[0]] = 1  # Mark the points in the grid

plt.plot([2, 12], [2, 13], color='red')
plt.plot([2,13], [13,2], color='green')

plt.grid(True)
#                                     l, r,  b, t
plt.imshow(grid, cmap='gray', extent=[0, 15, 0, 15])
plt.xticks(np.arange(0, 16, 1))
plt.yticks(np.arange(0, 16, 1))

plt.title('Bresenham\'s Line Algorithm')
plt.show()
```
