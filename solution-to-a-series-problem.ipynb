```python
import numpy as np
import matplotlib.pyplot as plt

# Define a range of n values
n_values = np.arange(1, 100)  # from 1 to 99

# Calculate the left side of the equation for each n
left_side = (np.arange(1, n_values.max()) / n_values[:, None]).sum(axis=1) + 1

print([zip(n_values, left_side)])

# Define the right side of the equation
right_side = 37 / 2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, left_side, label='Left Side: (1/n + 2/n + ... + (n-1)/n + 1)', color='blue')
plt.axhline(y=right_side, color='red', linestyle='--', label='Right Side: 37/2')

# Highlighting the point where left side equals right side
plt.scatter(36, right_side, color='green')  # This is where n=36
plt.text(36, right_side, ' n=36', fontsize=12, verticalalignment='bottom')

# Adding labels and title
plt.title('Comparison of Left Side and Right Side of the Equation')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.ylim(0, 25)
plt.xlim(1, 100)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x_values = np.arange(1, 16)  # From 1 to 15

# Calculate f(x) for each x value
f_values = x_values**2 + x_values + 1

# Create a line plot
plt.figure(figsize=(12, 6))

# Line plot
plt.subplot(1, 2, 1)
plt.plot(x_values, f_values, marker='o', color='blue', linestyle='-', label='f(x) = x^2 + x + 1')
plt.title('Line Plot of f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xticks(x_values)  # Set x ticks to integer values
plt.grid()
plt.legend()

# Create a scatter plot
plt.subplot(1, 2, 2)
plt.scatter(x_values, f_values, color='red', label='f(x) values')
plt.title('Scatter Plot of f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xticks(x_values)  # Set x ticks to integer values
plt.grid()
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
